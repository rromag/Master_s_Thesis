```python
def get_docs_for_topic(topic_id, train_docs, topics):
    """
    Return all training documents that belong to a given topic.

    Parameters
    ----------
    topic_id : int
        The topic you want to extract documents for.
    train_docs : list of str
        The training documents used in BERTopic.
    topics : array-like of int
        Topic assignment for each training document.

    Returns
    -------
    list of str
        Documents belonging to the specified topic.
    """
    return [doc for doc, t in zip(train_docs, topics) if t == topic_id]

print(get_docs_for_topic(84, train_docs, topics))
```
